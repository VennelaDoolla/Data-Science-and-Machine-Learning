```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Aerofit.csv')
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Product'].value_counts()
```

```python
tm1=df[df.Product=='KP281']
tm2=df[df.Product=='KP481']
tm3=df[df.Product=='KP781']
```

```python
tm1.describe()
```

```python
tm2.describe()
```

```python
tm3.describe()
```

```python
tm=[tm1['Income'],tm2['Income'],tm3['Income']]
```

```python
plt.figure(figsize=(10,10))
# Creating plot
plt.boxplot(tm)
plt.xlabel('Products')
plt.ylabel('Income')
plt.title('Distribution of Incomes of People who buy Products 1,2 and 3')
 
# show plot
plt.show()
```

```python
plt.figure(figsize=(10,10))
# Creating plot
plt.boxplot([tm1['Fitness'],tm2['Fitness'],tm3['Fitness']])
plt.ylabel('Fitness')
plt.xlabel('Products')

plt.title('Distribution of Fitness of People who buy Products 1,2 and 3')
plt.show()
```

```python
plt.figure(figsize=(10,10))
# Creating plot
plt.scatter(df['Fitness'],df['Income'])
plt.xlabel('Fitness')

plt.ylabel('Income in $')
plt.title('Distribution of Incomes of People respective to fitness')
# show plot
plt.show()
```

```python
plt.figure(figsize=(10,10))
# Creating plot
plt.scatter(df['Fitness'],df['Usage'])
plt.xlabel('Fitness')
plt.ylabel('usage in hours')
plt.title('Distribution of usage of treadmill in hours respective to fitness')
# show plot
plt.show()
```
